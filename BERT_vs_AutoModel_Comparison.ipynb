```python
# 1. PyTorch AutoTokenizer and AutoModel
from transformers import AutoTokenizer, AutoModel

print("=== AutoTokenizer and AutoModel (PyTorch) ===")
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

text = "Transformers are awesome!"

tokens = tokenizer.tokenize(text)
print("AutoTokenizer tokens:", tokens)

inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)
print("AutoModel output shape:", outputs.last_hidden_state.shape)
print(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0]))
```
